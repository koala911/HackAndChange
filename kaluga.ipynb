```python
import numpy as np

def replace_binary_value(b):
    if b and pd.notnull(b):
        return 1
    else:
        return -1
```

```python
data.iloc[:, 12:16] = data.iloc[:, 12:16].agg(lambda row: row.agg(replace_binary_value, axis=0), axis=1)
```

```python
target_names = ['Postamat_daily', 'cashbox_daily']
feature_names = ['Postamat_trend', 'cashbox_MAX', 'cashbox_trend', 'hh_500', 'value1', 'value2', 'NEAR_Malls', 'NEAR_Business_centers', 'NEAR_metro_rjd', 'NEAR_Stations']

plt.figure(figsize=(20, 70))
plot_number = 0
for target_name in target_names:
    for feature_name in feature_names:
        plot_number += 1
        plt.subplot(10, 2, plot_number)
        plt.scatter(data_kaluga[feature_name], data_kaluga[target_name])
        plt.xlabel(feature_name)
        plt.ylabel(target_name)
```
